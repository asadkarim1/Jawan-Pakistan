## ***6. Fetch Random Sample From the Dataset (50%)***

```python
# Fetch a random sample from the dataset (50%)
sample_df = df.sample(frac=0.5, random_state=1)
sample_df.head()
```

## ***7. Perform Data Cleaning [Replace'?' with NaN]***

```python
sample_df.replace('?', np.nan, inplace=True)
sample_df
```

## ***8. Drop All The Missing Values***

```python
sample_df.dropna(inplace=True)
sample_df
```

## ***9. Check Duplicate Data and Drop Them***

```python
duplicates = sample_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
sample_df.drop_duplicates(inplace=True)
```

## ***10. Get Overall Statistics About The Dataframe***

```python
sample_df.describe()
```

## ***11. Drop The Columns education-num, capital-gain and capital-loss***


```python
sample_df.drop(columns=['educational-num', 'capital-gain', 'capital-loss'], axis=1, inplace=True)
sample_df
```

## ***12. What Is The Distribution of Age Column Show In Graph?***


```python
plt.figure(figsize=(10, 6))
sns.histplot(sample_df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.savefig('age_distribution.png')
plt.show()
```

## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
# Find total number of persons having age between 17 to 48 (inclusive) using the 'between' method
C_B = sample_df[sample_df['age'].between(17, 48)].shape[0]

# Find total number of persons having age between 17 to 48 (inclusive) without using the 'between' method
c_w_b = sample_df[(sample_df['age'] >= 17) & (sample_df['age'] <= 48)].shape[0]

C_B, c_w_b
```

## ***14. What is The Distribution of Workclass Column***

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=sample_df, x='workclass', order=df['workclass'].value_counts().index)
plt.title('Distribution of Workclass')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## ***16. Create A Boxplot Of The "Income" Column Against The "Age" Column, Segmented By Gender***

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=sample_df, x='age', y='income', hue='gender')
plt.title('Boxplot of Income Against Age by Gender')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
sample_df['income'] = sample_df['income'].replace({'<=50K': 0, '>50K': 1})

# Display the updated income column to verify changes
income_updated = sample_df['income'].value_counts()
print(income_updated)
```

## ***18. Which Workclass getting The Highest Salary***

```python
# Find the workclass that gets the highest salary
highest_salary_workclass = sample_df.groupby('workclass')['income'].mean().idxmax()
print(f"Workclass with the highest average salary: {highest_salary_workclass}")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
female = sample_df[sample_df['gender'] == 'Female']['income'].mean()
male = sample_df[sample_df['gender'] == 'Male']['income'].mean()

print(f"Chance of getting salary >50K (Female): {female * 100:.2f}%")
print(f"Chance of getting salary >50K (Male): {male* 100:.2f}%")
```
